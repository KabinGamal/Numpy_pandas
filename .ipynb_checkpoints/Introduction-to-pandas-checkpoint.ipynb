1. series (1D)
2. DataFrame (2D)

For the series

```python
cars=pd.Series(["BMW","Mercedes","Porsche"])
cars
```

```python
color=pd.Series(["Grey","Black","Golden","Red"])
color
```

For the dataFrame

```python
dataframe=pd.DataFrame({"Cars":cars,"Color":color})
```
