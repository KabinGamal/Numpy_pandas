# Plotting multiple digaram at once

```python
fig ,((ax1,ax2),(ax3,ax4)) =plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5)
)
ax1.plot(kabs,sabs),
ax2.bar(kabs,sabs),
ax3.scatter(kabs,sabs)
ax4.hist(np.random.randn(100))
plt.show()
```
